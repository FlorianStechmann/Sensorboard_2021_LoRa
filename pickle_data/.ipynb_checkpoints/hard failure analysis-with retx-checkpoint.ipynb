```python
### PRR of the primary board (PB)
num_pkts_sent_pb = 0 
num_pkts_rx_pb = 0
packet_num_pb = []
packet_num_conf_pb = []

num_pkts_sent_rb = 0 
num_pkts_rx_rb = 0
packet_num_rb = []

#### split the primary board in 2 sections where board restarted
num_pb = []
conf_pb = []
for d in data[1][:break_ind]:  ##pb
    if d[2] not in packet_num_pb:
        num_pb += [d[2]]
        if d[5]-d[4]<10:
            conf_pb += [d[2]]

packet_num_pb += [num_pb]
packet_num_conf_pb += [conf_pb]
            
#### split the primary board in 2 sections where board restarted
num_pb = []
conf_pb = []
for d in data[1][break_ind:]:  ##pb
    if d[2] not in packet_num_pb:
        num_pb += [d[2]]
        if d[5]-d[4]<10:
            conf_pb += [d[2]]

packet_num_pb += [num_pb]
packet_num_conf_pb += [conf_pb]
####################
        
for d in data[3]:   #rb
    if d[2] > first_faulty_pkt or d[2] <= last_faulty_pkt: ### take only packets during faulty data
        if d[0][0] != -1:    # filter out hb
            packet_num_rb += [d[2]]

#### total
num_pkts_rx_pb = len(packet_num_pb)
prr_pb = (num_pkts_rx_pb/total_pkts) * 100    ### packets rx from only PB out of total packets sent by node
print('PRR of the primary board:', prr_pb)

num_pkts_rx_rb = len(packet_num_rb)
prr_sn = (num_pkts_rx_pb+num_pkts_rx_rb)/total_pkts * 100
print('PRR of the sensor node:', prr_sn)

#### with conf int.
num_pkts_rx_pb = len(packet_num_conf_pb)
prr_pb = (num_pkts_rx_pb/total_pkts) * 100    ### packets rx from only PB out of total packets sent by node
print('PRR of the primary board:', prr_pb)

num_pkts_rx_rb = len(packet_num_rb)
prr_sn = (num_pkts_rx_pb+num_pkts_rx_rb)/total_pkts * 100
print('PRR of the sensor node:', prr_sn)
```
